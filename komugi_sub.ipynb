```python
import itertools
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## 【問題1】2×2マスのチェス盤の小麦の数

```python
def seq(i = 1):
    while True:
        yield i
        i *= 2

arr = np.array(list(itertools.islice(seq(), 2 * 2)))
board = arr.reshape((2, 2))
board
```

## 【問題2】n×mマスへの拡張

```python
def chess_board(size):
    n, m = size
    arr = np.array(list(itertools.islice(seq(), n * m)))
    return arr.reshape(size)

board = chess_board((8, 8))
board
```

## 【問題3】小麦の数の合計

```python
s = np.sum(board)
print(f"8×8マスのチェス盤の小麦の数の合計: {s}")
```

```python
m = np.mean(board, axis=0)

plt.xlabel("column")
plt.ylabel("number")
plt.title("number in each column")
plt.bar(range(len(m)), m)
plt.show()
```

## 【問題4】小麦の数のヒートマップ

```python
plt.xlabel("column")
plt.ylabel("row")
plt.title("heatmap")
plt.pcolor(board)
plt.gca().invert_yaxis()
plt.show()
```

## 【問題5】後半は前半の何倍か

```python
former = board[0:4,:]
latter = board[4:8,:]

half1 = np.sum(former)
half2 = np.sum(latter)

print(f"チェス盤の後半にある小麦の数は前半にある小麦の数の {half2 / half1} 倍")
```

## 【問題6】他の計算方法によるn×mマスへの拡張

```python
def chess_board_append(size):
    n, m = size
    arr = np.array([1.0])
    for _ in range(n * m - 1):
        arr = np.append(arr, 2 * arr[-1])
    return arr.reshape(size)

board_append = chess_board_append((8, 8))
board_append
```

```python
def chess_board_broadcast(size):
    n, m = size
    return 2.0 ** np.arange(n * m).reshape(size)

board_broadcast = chess_board_broadcast((8, 8))
board_broadcast
```

## 【問題7】計算時間の比較

```python
%%timeit
board = chess_board((8, 8))
```

```python
%%timeit
board_append = chess_board_append((8, 8))
```

```python
%%timeit
board_broadcast = chess_board_broadcast((8, 8))
```

append を使った方法が一番遅い。
新しい配列コピーがいちいち作られるためだと思われる。

numpy のブロードキャストを使うと一番早いが、最初の素朴な方法もこれに比べて極端に遅いとは言えない。
